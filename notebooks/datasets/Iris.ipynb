# Iris flower data set

Quote from [Iris flower data set](https://en.wikipedia.org/wiki/Iris_flower_data_set) Wikipedia page:

> The ***Iris* flower data set** or **Fisher's *Iris* data set** is a multivariate data set introduced by the British statistician, eugenicist, and biologist Ronald Fisher in his 1936 paper *The use of multiple measurements in taxonomic problems* as an example of linear discriminant analysis.

For more details, please refer to [Iris flower data set](https://en.wikipedia.org/wiki/Iris_flower_data_set)

This notebook is to show how to load *Iris* data set by using `scikit-learn` package and to visualize it by using `seaborn` 

```python
from sklearn import datasets
import pandas as pd

# Load Iris data set from sklearn.datasets
iris = datasets.load_iris()

# Convert the above Iris data set (in dictionary) into Pandas dataframe
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Series([iris.target_names[x] for x in iris.target])

iris_df.head()
```

```python
import seaborn as sns

# Draw a pair plot from Iris data set with all 4 features
sns.pairplot(iris_df, hue='species')
```
