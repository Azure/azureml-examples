# Use OpenAI SDK with Stability AI models in Azure AI and Azure ML for Text to Image requests

Use openai SDK to consume Stability AI deployments in Azure AI and Azure ML. The Stability family of models in Azure AI and Azure ML offers an API compatible with the OpenAI Images Generation API. 

The example below shows how to make this transition using the OpenAI Python Library.

## Prerequisites

Before we start, there are certain steps we need to take to deploy the models:

* Register for a valid Azure account with subscription

* Make sure you have access to Azure AI Studio

* Create a project and resource group

* Select Stability AI models from Model catalog. This example assumes you are Stable Diffusion 3.5 Large.

* Deploy with "Pay-as-you-go"

Once deployed successfully, you should be assigned for an API endpoint and a security key for inference.

For more information, you should consult Azure's official documentation here for model deployment and inference.

To complete this tutorial, you will need to:

* Install openai:
    
    pip install openai

* Install pillow:
    
    pip install pillow

* Install matplotlib
    
    pip install matplotlib

## Example
The following is an example about how to use openai with a Stability AI Text to Image model deployed in Azure AI and Azure ML:



```python
import base64
from openai import OpenAI
from typing import Any
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
def display_base64_image(base64_string):
    """Decode a base64 string and display it as an image."""
    image_data = base64.b64decode(base64_string)
    image = Image.open(BytesIO(image_data))
    plt.imshow(image)
    plt.axis("off")
    plt.show()
```

You will need to have a Endpoint url and Authentication Key associated with that endpoint. This can be acquired from previous steps. To work with openai, configure the client as follows:

base_url: Use the endpoint URL from your deployment.
api_key: Use your API key.

```python
client = OpenAI(base_url="", api_key="")
```

Use the client to create image generation requests:

```python
response = client.images.generate(
    prompt="A painting of a sunset over a mountain lake", size="1024x1024"
)
```

The generated image can be accessed as follows:

```python
display_base64_image(response.image)
```
