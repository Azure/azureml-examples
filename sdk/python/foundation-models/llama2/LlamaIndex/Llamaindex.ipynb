```python
from langchain.chat_models import ChatAnyscale, ChatOpenAI
from llama_index.llms.langchain import LangChainLLM
from llama_index.core import PromptTemplate
from dotenv import load_dotenv
import os
```

LangChain LLM

Pass the openai base and url key for the Model and suffix it with "/v1"

```python
load_dotenv("config.env")
api_key=os.getenv("api_key")
api_base=os.getenv("api_base")
```

```python
llm = LangChainLLM(ChatOpenAI( openai_api_base=api_base,
    openai_api_key=api_key))
#pass the openai base and url key for the Model and suffix it with "/v1"
```

Giving the prompt

```python
stream = await llm.astream(PromptTemplate("Hi, write a short story"))
```

Output

```python
async for token in stream:
    print(token, end="")
```
